## Background
***

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

```python
# Code to calculate mean of a list of numbers:

num_list = [1,4,5,6,8,9,12,15,18,20,25,28,30] # define a list of numbers for testing

def find_mean(lst):
    """Returns the mean of a list of numbers"""
    return (sum(lst)/ len(lst))
     

find_mean(num_list) # test the function and check if it works!
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code to calculate standard deviation of a list of numbers:

num_list = [1,4,5,6,8,9,12,15,18,20,25,28,30] # define a list of numbers for testing

def find_std(lst):
    """Returns the standard deviation of a list of numbers"""
    mean = find_mean(lst)
    total = 0
    for num in lst:
        total += (num - mean) ** 2
    return ((total/len(lst)) ** 0.5)

find_std(num_list) # test the function and check if it works!
```

### Histograms
***
The histograms plotted below will help to provide a better understanding of the underlying distributions of the participation rates and scores of the SAT and ACT tests as well as the number of test takers and scores for the intended majors for SAT test. 

```python
def plot_hist(df, col_lst, title_lst):
    """custom function for plotting of histograms"""
    fig = plt.figure(figsize = (18,7))
    fig.subplots_adjust(hspace=0.3, wspace=0.2)
    
    for ind, col in enumerate(col_lst):
        ax = fig.add_subplot(2,3, (ind + 1))
        ax.hist(df[col], bins = 15, edgecolor='white')
        ax.set_title(title_lst[ind])
        
```

```python
# Histograms for participation rates for SAT and ACT

part_col = ['sat_participation_2017', 'sat_participation_2018', 'sat_participation_2019', 'act_participation_2017', 'act_participation_2018', 'act_participation_2019']
part_title = ['SAT Participation (2017)', 'SAT Participation (2018)', 'SAT Participation (2019)', 'ACT Participation (2017)', 'ACT Participation (2018)', 'ACT Participation (2019)']

plot_hist(scores_final_merged, part_col, part_title);
```

From the histograms of the participation rates of SAT test, it shows that there are more states at the extreme ends of the participation rates. As for the ACT test, the peaks are at the higher ends of the participation rates and not at the lower end. With the peaks of both SAT and ACT test at the ends of the range which is near 0% and 100%, this would mean that the distribution of the participation rates is not normally distributed. 

```python
# Histograms for total scores for SAT and ACT

scores_col = ['total_score_2017', 'total_score_2018', 'total_score_2019', 'composite_score_2017', 'composite_score_2018', 'composite_score_2019']
scores_title = ['SAT Score (2017)', 'SAT Score (2018)', 'SAT Score (2019)', 'ACT Score (2017)', 'ACT Score (2018)', 'ACT Score (2019)']

plot_hist(scores_final_merged, scores_col, scores_title);
```

From the histograms of the scores of the SAT test and ACT test, the trends are generally similar with the two main peaks being observed across the histogram. The distribution seems to be bimodal and not normal distributed. The peaks of the SAT Score is observed at around 1050 to 1100 across the years and for ACT Score is observed at around 20 for 2017 and 2018 and 24 for 2019. 

```python
# Histograms for scores and number of test_takers for intended majors for SAT

majors_col = ['test_takers_2017', 'test_takers_2018', 'test_takers_2019', 'total_2017', 'total_2018', 'total_2019']
majors_title = ['No. of test takers (2017)', 'No. of test takers (2018)', 'No. of test takers (2019)', 'Score for majors (2017)', 'Score for majors (2018)', 'Score for majors (2019)']

plot_hist(sat_majors_scores_merged, majors_col, majors_title);
```

## Additional Insights

### West Virginia

As observed from the bar charts, it can be seen that West Virginia has a massive increase in SAT participation rate from 2018 to 2019. Checking the dataframe, we can see that the increase in SAT participation rate was from 28% in 2018 to 99% in 2019.
